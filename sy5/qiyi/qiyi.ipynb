```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.linalg import *


```

```python
from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data
y = iris.target
x.shape
```

```python
from scipy.linalg import svd  

# 假设 x 是已定义的矩阵
u, s, v = svd(x)
print(u,s,v)
```
